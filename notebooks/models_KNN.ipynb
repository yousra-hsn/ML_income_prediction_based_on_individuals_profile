## Regrouping features

### Values not referenced

```python
data['workclass'] = data['workclass'].replace({'?': 'Not referenced'})

data['native-country'] = data['native-country'].replace({'?': 'Not referenced'})

data['occupation'] = data['occupation'].replace({'?': 'Not referenced'})
```

### Workclass

```python
# Regroup 'Without-pay' and 'Never-worked' to 'No revenu'
data['workclass'] = data['workclass'].replace({'Without-pay': 'No revenu', 'Never-worked': 'No revenu'})

# Regrouper 'Self-emp-not-inc' and 'Self-emp-inc' to 'Self-emp'
data['workclass'] = data['workclass'].replace({'Self-emp-not-inc': 'Self-emp', 'Self-emp-inc': 'Self-emp'})
```

### Marital-status

```python
data['marital-status'] = data['marital-status'].replace({'Divorced': 'Now Single', 'Separated': 'Now Single', 'Widowed': 'Now Single'})

data['marital-status'] = data['marital-status'].replace({'Married-civ-spouse': 'Married', 'Married-AF-spouse': 'Married'})
```

```python
data['relationship'] = data['relationship'].replace({'Husband': 'Married', 'Wife': 'Married'})
```

On met dans la catégorie "Other" tous les pays qui obtiennent un nombre d'observations inférieur à 200. 

```python
# Counting for each value of the variable 'native-country' the number of individuals who have this value
filtered = data[data['native-country'] != 'United-States']

for country in filtered['native-country'].unique():
    nb_samples_associated = filtered[filtered['native-country'] == country].shape[0]
    if nb_samples_associated < 200:
        data['native-country'] = data['native-country'].replace({country: 'Other'})
        
```

## Modelisation - KNN

```python
X_train = np.ascontiguousarray(X_train_encoded) # Converts the array to C-contiguous
X_test = np.ascontiguousarray(X_test_encoded) # Converts the array to C-contiguous
```

```python
X_train.flags.c_contiguous
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()

param_grid = {
    'n_neighbors': list(range(1, 11, 2)) + list(range(11, 81, 10)),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

best_knn = grid_search.best_estimator_
print("Best parameters for k-NN:", grid_search.best_params_)
```

```python
def visualize_knn(grid_search_):
    # Extract cv_results_ and convert to DataFrame
    results = pd.DataFrame(grid_search_.cv_results_)
    # Convert hyperparameters to appropriate data types if needed
    results['param_n_neighbors'] = results['param_n_neighbors'].astype(int)

    # Visualize mean test scores for different metrics and weights
    plt.figure(figsize=(14, 8))
    sns.lineplot(
        data=results,
        x='param_n_neighbors',
        y='mean_test_score',
        hue='param_metric',
        style='param_weights',
        markers=True,
        dashes=False
    )
    plt.title('Mean Test Scores by n_neighbors, Metric, and Weights')
    plt.xlabel('Number of Neighbors (n_neighbors)')
    plt.ylabel('Mean Test Score')
    plt.legend(title='Metric and Weights')
    plt.show()

    # Create a combined column for metric and weight
    results['metric_weight'] = results['param_metric'].astype(str) + "_" + results['param_weights'].astype(str)

    # Create a pivot table to visualize results as a heatmap
    pivot_table = results.pivot_table(
        values='mean_test_score',
        index='param_n_neighbors',
        columns='metric_weight',
        aggfunc='mean'
    )

    # Plot heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(pivot_table, annot=True, cmap='viridis')
    plt.title('Heatmap of Mean Test Scores')
    plt.xlabel('Metric and Weights')
    plt.ylabel('Number of Neighbors (n_neighbors)')
    plt.show()
    
visualize_knn(grid_search)
```

```python
# Suivant un premier grid_search on faire un deuxième en résuisant les params

param_grid = {
    'n_neighbors': range(40, 70),
    'weights': ['uniform'],
    'metric': ['euclidean', 'minkowski']
}

grid_search_2 = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
grid_search_2.fit(X_train, y_train)

best_knn = grid_search_2.best_estimator_
print("Best parameters for k-NN:", grid_search_2.best_params_)
```

```python
visualize_knn(grid_search_2)
```
